```python
import pymongo as mongo
import pandas as pd
```

```python
# load the dataset
def load_dataset() -> pd.DataFrame:
    """
    Load the dataset from the CSV file and return it as a pandas DataFrame.

    Args:
        None

    Returns:
        pd.DataFrame: The dataset loaded from the CSV file.
    """

    data_path = "https://github.com/donadviser/datasets/raw/master/data-don/marketing_campaign.zip"
    data = pd.read_csv(data_path, compression='zip', sep="\t")
    return data
```

```python
data_raw = load_dataset()
```

```python
#convert the dataframe to dict data structure for >> MongoDB
data_dict = data_raw.to_dict(orient="records")
```

```python
DB_NAME = "marketing_campaign" # Specify the Database Name

# Connection URL
CONNECTION_URL = "mongodb+srv://donadviser:password@cluster0.yadqc.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0"


# Establish a connection with mongoDB
client = mongo.MongoClient(CONNECTION_URL)

# Create a database
dataBase = client[DB_NAME]

# Create a collection
COLLECTION_NAME = "customer_segmentation"
collection = dataBase[COLLECTION_NAME]

# insert above records in the collection
rec = collection.insert_many(data_dict)

# Let's verify all the records at once present in the record with all the field
all_records = collection.find()

for idx, record in enumerate(all_records):
    print(f"{idx}: {record}")
```
